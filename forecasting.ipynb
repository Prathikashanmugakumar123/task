```python
import numpy as np
from statsmodels.tsa.arima.model import ARIMA

# Stock price data
data = [100, 102, 105, 107, 110]

# Create ARIMA model
model = ARIMA(data, order=(1, 1, 1))
model_fit = model.fit()

# Forecast next day price
forecast = model_fit.forecast(steps=1)

print("Forecasted stock price for next day:", forecast[0])
```
