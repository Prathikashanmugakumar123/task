```python
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LinearRegression
# Data
X = np.arange(11).reshape(-1,1)   # dummy feature (time index)
y = np.array([0,0,0,1,2,0,3,0,1,0,2])

# Classification (Zero or Not)
y_class = (y > 0).astype(int)   # convert to 0 or 1

clf = LogisticRegression()
clf.fit(X, y_class)
# Step 2 â†’ Regression (Only non-zero values)
X_nonzero = X[y > 0]
y_nonzero = y[y > 0]

reg = LinearRegression()
reg.fit(X_nonzero, y_nonzero)
# Prediction
x_new = [[12]]
is_nonzero = clf.predict(x_new) 
if is_nonzero == 0:
    print("Prediction:", 0)
else:
    count = reg.predict(x_new)
    print("Prediction:", int(count[0]))
```
